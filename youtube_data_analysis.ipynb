```python
k1 = widgets.Dropdown(
     options=video['genre'].unique(),
     value='Music',
     description='Genre',
     disabled=False,
  )

def Genre(m):
  gen = video[video['genre']==m]
  k2 = widgets.Dropdown(
      options=gen['channelTitle'].unique(),
      value=gen['channelTitle'].unique()[0],
      description='channel',
      disabled=False,
 )
  def channel(d):
    ht = gen.sort_values(by = 'view_count',ascending= False).head()
    sns.barplot(y='title',x='view_count',data=ht)
    plt.title(d)
    plt.xlabel('view')
    plt.ylabel('video Name')
    plt.show()

  widgets.interact(channel,d=k2)

```

```python
widgets.interact(Genre,m=k1)
```

```python
r1 = widgets.Dropdown(
     options=video['channelTitle'].unique(),
     value=video['channelTitle'].unique()[0],
     description='Channel',
     disabled=False,
  )

def Channel(m):
  gen = data[data['channelTitle']==m]
  r2 = widgets.Dropdown(
      options=gen['title'].unique(),
      value=gen['title'].unique()[0],
      description='video',
      disabled=False,
 )
  def video(d):
    ht = gen[gen['title']==d]
    sns.lineplot(y='trending_date',x='view_count',data=ht)
    plt.title(d)
    plt.xlabel('view')
    plt.ylabel('date')
    plt.show()

  widgets.interact(video,d = r2)
```
